```python
#get sinopsis
import tmdbsimple as tmdb

tmdb.API_KEY = 'ed5e66b6e5d9db5d19e3a127aafcb3f5'

#GPT
def get_synopsis(imdb_id, title_type):
    try:
        search = tmdb.Find(imdb_id)
        response = search.info(external_source='imdb_id')
        if title_type == 'movie' and 'movie_results' in response:
            results = response['movie_results']
            if results:
                movie_id = results[0]['id']
                movie = tmdb.Movies(movie_id)
                return movie.info()['overview']
        elif title_type == 'tvSeries' and 'tv_results' in response:
            results = response['tv_results']
            if results:
                series_id = results[0]['id']
                series = tmdb.TV(series_id)
                return series.info()['overview']
    except Exception as e:
        print(f"Error al obtener la sinopsis para IMDb ID {imdb_id}: {e}")
    return None
```

```python
#predict emotion
from transformers import pipeline

pipe = pipeline("text-classification", model="TFMUNIR/distilbert-base-uncased-finetuned-emotion-movies-186k")

label_meaning = {
    'LABEL_0': 'sadness',
    'LABEL_1': 'joy',
    'LABEL_2': 'love',
    'LABEL_3': 'anger',
    'LABEL_4': 'fear',
    'LABEL_5': 'surprise',
}

def getEmotion(text):
    labels = pipe(text)
    if len(labels):
        if 'label' in labels[0]:
            return label_meaning[labels[0]['label']]#asume que esta en el primer lugar
    return 'None'
```

```python
#plataformas
def getPlataformas(movie):
    # Obtener informaciÃ³n de las plataformas de transmisiÃ³n
    streaming_info = movie.watch_providers()
    streaming_data = streaming_info.get('results', {})
    link = ""

    #print(streaming_data)

    # Filtrar las plataformas de acuerdo a la regiÃ³n y "flatrate"
    platforms = [] #no existe un link directo, solo extraera los nombres
    
    for region, data in streaming_data.items():
        #print("region ", region)
        if region == "US" and "flatrate" in data:#cambiar a peru
            flatrate_info = data["flatrate"]

            for plataforma in flatrate_info:
                platforms.append(plataforma['provider_name'])

            link = data["link"]
            return platforms,link, True
            
    return platforms,link, False
```
